```python
import sys
from PyQt5.QtWidgets import (
    QApplication, QWidget, QVBoxLayout, QHBoxLayout, QLabel, QPushButton,
    QListWidget, QCheckBox, QScrollArea, QFrame
)

#courses from "UniTag Formatted Dictionary" doc go here 
courses = {
    "CCAS 0920": {
        "Department": "Columbian College",
        "CRN": "30742",
        "Subject": "CCAS",
        "Course Number": "0920",
        "Course Name": "Continuing Research - Master's",
        "Tags": ["Columbian College"]
    },
    "CCAS 0940": {
        "Department": "Columbian College",
        "CRN": "31705",
        "Subject": "CCAS",
        "Course Number": "0940",
        "Course Name": "Continuing Research - Doctoral",
        "Tags": ["Columbian College"]
    },
    "CCAS 1001": {
        "Department": "Columbian College",
        "CRN": "34426",
        "Subject": "CCAS",
        "Course Number": "1001",
        "Course Name": "First-Year Experience",
        "Tags": ["Columbian College"]
    },
    "CLAS 2105": {
        "Department": "Classical Studies",
        "CRN": "33307",
        "Subject": "CLAS",
        "Course Number": "2105",
        "Course Name": "Politics&Power:Roman Republic",
        "Tags": ["Classical Studies"]
    },
    "HIST 2001": {
        "Department": "Classical Studies",
        "CRN": "35676",
        "Subject": "HIST",
        "Course Number": "2001",
        "Course Name": "Politics&Power:Roman Republic",
        "Tags": ["Classical Studies"]
    },
    "CLAS 2106": {
        "Department": "Classical Studies",
        "CRN": "37072",
        "Subject": "CLAS",
        "Course Number": "2106",
        "Course Name": "Mythology of the Classical World",
        "Tags": ["Classical Studies"]
    },
    "CLAS 2112": {
        "Department": "Classical Studies",
        "CRN": "31753",
        "Subject": "CLAS",
        "Course Number": "2112",
        "Course Name": "History of Ancient Greece",
        "Tags": ["Classical Studies"]
    },
    "HIST 2112": {
        "Department": "Classical Studies",
        "CRN": "31752",
        "Subject": "HIST",
        "Course Number": "2112",
        "Course Name": "History of Ancient Greece",
        "Tags": ["Classical Studies"]
    },
    "CPJ 2091": {
        "Department": "Corcoran Photojournalism",
        "CRN": "34142",
        "Subject": "CPJ",
        "Course Number": "2091",
        "Course Name": "Photojournalism Studio Sequencing and Narrative Strategies",
        "Tags": ["Photojournalism"]
    },
    "CPJ 3090": {
        "Department": "Corcoran Photojournalism",
        "CRN": "32091",
        "Subject": "CPJ",
        "Course Number": "3090",
        "Course Name": "Photojournalism Studio Visual Reportage",
        "Tags": ["Photojournalism"]
    },
    "CSA 3501": {
        "Department": "Corcoran Photojournalism",
        "CRN": "34490",
        "Subject": "CSA",
        "Course Number": "3501",
        "Course Name": "Documentary Photography",
        "Tags": ["Photojournalism"]
    },
    "CPJ 4085": {
        "Department": "Corcoran Photojournalism",
        "CRN": "36090",
        "Subject": "CPJ",
        "Course Number": "4085",
        "Course Name": "Photojournalism Directed Study",
        "Tags": ["Photojournalism"]
    },
    "CPJ 4090": {
        "Department": "Corcoran Photojournalism",
        "CRN": "32090",
        "Subject": "CPJ",
        "Course Number": "4090",
        "Course Name": "Photojournalism Thesis I",
        "Tags": ["Photojournalism"]
    },
    "CMUS 1063": {
        "Department": "Corcoran Music",
        "CRN": "35055",
        "Subject": "CMUS",
        "Course Number": "1063",
        "Course Name": "Chamber Ensembles",
        "Tags": ["Music"]
    },
    "CMUS 1071": {
        "Department": "Corcoran Music",
        "CRN": "35170",
        "Subject": "CMUS",
        "Course Number": "1071",
        "Course Name": "Jam Session",
        "Tags": ["Music"]
    }, 
    "TSAP 6000":{
        "Department":"Strategic Mgt & Public Policy",
        "CRN":"33254",
        "Subject":"TSAP",
        "Course Number":"6000",
        "Course Name":"Teaching in the Sciences/profs ",
        "Tags":["Strategic Mgt & Public Policy"]
    },
    "SMPP 4900":{
        "Department":"Strategic Mgt & Public Policy",
        "CRN":"37902",
        "Subject":"SMPP",
        "Course Number":"4900",
        "Course Name":"Measuring Sustainability ",
        "Tags":["Science"]
    },
	"SMPP 4995":{
        "Department":"Strategic Mgt & Public Policy",
        "CRN":"33278",
        "Subject":"SMPP",
        "Course Number":"4995",
        "Course Name":"Independent Study ",
        "Tags":["Strategic Mgt & Public Policy"]
    },
	"SMPP 6202":{
        "Department":"Strategic Mgt & Public Policy",
        "CRN":"37920",
        "Subject":"SMPP",
        "Course Number":"6202",
        "Course Name":"Business-Government Relations ",
        "Tags":["Strategic Mgt & Public Policy"]
    },
	"SMPP 6204":{
        "Department":"Strategic Mgt & Public Policy",
        "CRN":"36045",
        "Subject":"SMPP",
        "Course Number":"6204",
        "Course Name":"Strategy Concepts ",
        "Tags":["Strategic Mgt & Public Policy"]
    },
    "SMPP 6298":{
        "Department":"Strategic Mgt & Public Policy",
        "CRN":"30522",
        "Subject":"SMPP",
        "Course Number":"6298",
        "Course Name":"Strategy Concepts ",
        "Tags":["Strategic Mgt & Public Policy"]
    },
    "SUST 1001":{
        "Department":"Sustainability",
        "CRN":"36440",
        "Subject":"SUST",
        "Course Number":"1001",
        "Course Name":"Introduction to Sustainability ",
        "Tags":["Sustainability"]
    },
    "SUST 2004":{
        "Department":"Sustainability",
        "CRN":"36449",
        "Subject":"SUST",
        "Course Number":"2004",
        "Course Name":"Introduction to Sustainability Communication ",
        "Tags":["Sustainability"]
    },
    "SUST 2005":{
        "Department":"Sustainability",
        "CRN":"38040",
        "Subject":"SUST",
        "Course Number":"2005",
        "Course Name":"Responsible Fashion ",
        "Tags":["Sustainability"]
    },
    "SUST 3093":{
        "Department":"Sustainability",
        "CRN":"37049",
        "Subject":"SUST",
        "Course Number":"3093",
        "Course Name":"Environmental Ethics & Justice ",
        "Tags":["Sustainability"]
    },
    "SUST 3096":{
        "Department":"Sustainability",
        "CRN":"36437",
        "Subject":"SUST",
        "Course Number":"3096",
        "Course Name":"Research in Sustainability ",
        "Tags":["Sustainability"]
    },
    "SPAN 1001":{
        "Department":"Spanish",
        "CRN":"35373",
        "Subject":"SPAN",
        "Course Number":"1001",
        "Course Name":"Elementary Spanish I",
        "Tags":["Spanish"]
    },
    "SPAN 1002":{
        "Department":"Spanish",
        "CRN":"35077",
        "Subject":"SPAN",
        "Course Number":"1002",
        "Course Name":"Elementary Spanish II",
        "Tags":["Spanish"]
    },
    "SPAN 1013":{
        "Department":"Spanish",
        "CRN":"32727",
        "Subject":"SPAN",
        "Course Number":"1013",
        "Course Name":"Intermediate Spanish I: the Spanish-Speaking World",
        "Tags":["Spanish"]
    },
    "SPAN 1014":{
        "Department":"Spanish",
        "CRN":"32729",
        "Subject":"SPAN",
        "Course Number":"1014",
        "Course Name":"Intermediate Spanish II:Through a Social Justice Lens",
        "Tags":["Spanish"]
    },
    "SPAN 2005":{
        "Department":"Spanish",
        "CRN":"30398",
        "Subject":"SPAN",
        "Course Number":"2005",
        "Course Name":"Advanced Spanish I",
        "Tags":["Spanish"]
    },
    "SOC 1001":{
        "Department":"Sociology",
        "CRN":"30396",
        "Subject":"SOC",
        "Course Number":"1001",
        "Course Name":"Introduction to Sociology",
        "Tags":["Sociology"]
    },
    "SOC 1002":{
        "Department":"Sociology",
        "CRN":"35952",
        "Subject":"SOC",
        "Course Number":"1002",
        "Course Name":"The Sociological Imagination",
        "Tags":["Sociology"]
    },
    "SOC 1003":{
        "Department":"Sociology",
        "CRN":"33068",
        "Subject":"SOC",
        "Course Number":"1003",
        "Course Name":"Introduction to Criminal Justice",
        "Tags":["Sociology"]
    },
    "SOC 2101":{
        "Department":"Sociology",
        "CRN":"30571",
        "Subject":"SOC",
        "Course Number":"2101",
        "Course Name":"Social Research Methods",
        "Tags":["Sociology"]
    },
    "SOC 2101":{
        "Department":"Sociology",
        "CRN":"32175",
        "Subject":"SOC",
        "Course Number":"2101",
        "Course Name":"Techniques of Data Analysis",
        "Tags":["Sociology"]
    },
    "SLHS 1011":{
        "Department":"Speech, Language & Hearing Sci",
        "CRN":"34115",
        "Subject":"SLHS",
        "Course Number":"1011",
        "Course Name":"The Art and Science of Effective Public Speaking",
        "Tags":["Speech, Language & Hearing Science"]
    },
    "SLHS 1071":{
        "Department":"Speech, Language & Hearing Sci",
        "CRN":"33909",
        "Subject":"SLHS",
        "Course Number":"1071",
        "Course Name":"Foundations of Human Communication",
        "Tags":["Speech, Language & Hearing Science"]
    },
    "SLHS 1071W":{
        "Department":"Speech, Language & Hearing Sci",
        "CRN":"34200",
        "Subject":"SLHS",
        "Course Number":"1071W",
        "Course Name":"Foundations of Human Communication",
        "Tags":["Speech, Language & Hearing Science"]
    },
    "SLHS 1072":{
        "Department":"Speech, Language & Hearing Sci",
        "CRN":"33911",
        "Subject":"SLHS",
        "Course Number":"1072",
        "Course Name":"Culturally Responsive Practices in Human Communication",
        "Tags":["Speech, Language & Hearing Science"]
    },
    "SLHS 1081":{
        "Department":"Speech, Language & Hearing Sci",
        "CRN":"35481",
        "Subject":"SLHS",
        "Course Number":"1081",
        "Course Name":"American Sign Language I",
        "Tags":["Speech, Language & Hearing Science"]
    },
    "SMPA 1000":{
        "Department":"School of Media&Public Affairs",
        "CRN":"31785",
        "Subject":"SMPA",
        "Course Number":"1000",
        "Course Name":"Hollywood & Politics",
        "Tags":["Media&Public Affairs"]
    },
    "SMPA 1225":{
        "Department":"School of Media&Public Affairs",
        "CRN":"31703",
        "Subject":"SMPA",
        "Course Number":"1226",
        "Course Name":"Forensics Practice (Debate)",
        "Tags":["Media&Public Affairs"]
    },
    "SMPA 2101":{
        "Department":"School of Media&Public Affairs",
        "CRN":"31148",
        "Subject":"SMPA",
        "Course Number":"2101",
        "Course Name":"Journalism: Theory and Practice",
        "Tags":["Media&Public Affairs"]
    },
    "SMPA 2102":{
        "Department":"School of Media&Public Affairs",
        "CRN":"31395",
        "Subject":"SMPA",
        "Course Number":"2102",
        "Course Name":"Intrduction fo Political Communication",
        "Tags":["Media&Public Affairs"]
    },
    "SMPA 2110W":{
        "Department":"School of Media&Public Affairs",
        "CRN":"32269",
        "Subject":"SMPA",
        "Course Number":"2110W",
        "Course Name":"Introduction to News Writing and Reporting",
        "Tags":["Media&Public Affairs"]
    },
    "SLAV 1001":{
        "Department":"Slavic Languages & Lit",
        "CRN":"37232",
        "Subject":"SLAV",
        "Course Number":"1001",
        "Course Name":"First-Year Russian I",
        "Tags":["Russian"]
    },
    "SLAV 1003":{
        "Department":"Slavic Languages & Lit",
        "CRN":"30392",
        "Subject":"SLAV",
        "Course Number":"1003",
        "Course Name":"Second-Year Russian I",
        "Tags":["Russian"]
    },
    "SLAV 1391":{
        "Department":"Slavic Languages & Lit",
        "CRN":"30395",
        "Subject":"SLAV",
        "Course Number":"1391",
        "Course Name":"Introduction to Russian Literature I",
        "Tags":["Russian"]
    },
    "SLAV 1391W":{
        "Department":"Slavic Languages & Lit",
        "CRN":"33277",
        "Subject":"SLAV",
        "Course Number":"1391W",
        "Course Name":"Introduction to Russian Literature I",
        "Tags":["Russian"]
    },
    "SLAV 2005":{
        "Department":"Slavic Languages & Lit",
        "CRN":"30394",
        "Subject":"SLAV",
        "Course Number":"2005",
        "Course Name":"Intermediate Russian I",
        "Tags":["Russian"]
    },
    "WGSS 1000":{
        "Department":"Women’s, Gender & Sexuality Study",
        "CRN":"36046",
        "Subject":"WGSS",
        "Course Number":"1000",
        "Course Name":"Islam in the Digital Age",
        "Tags":["Gender Culture"]
    },
    "REL 1000":{
        "Department":"Women’s, Gender & Sexuality Study",
        "CRN":"35769",
        "Subject":"REL",
        "Course Number":"1000",
        "Course Name":"Islam in the Digital Age",
        "Tags":["Gender Culture"]
    },
    "WGSS 1020":{
        "Department":"Women’s, Gender & Sexuality Study",
        "CRN":"36721",
        "Subject":"WGSS",
        "Course Number":"1020",
        "Course Name":"Approaches to Women’s History",
        "Tags":["Gender Culture"]
    },
    "HIST 1020":{
        "Department":"Women’s, Gender & Sexuality Study",
        "CRN":"36422",
        "Subject":"HIST",
        "Course Number":"1020",
        "Course Name":"Approaches to Women’s History",
        "Tags":["Gender Culture"]
    },
    "WGSS 2120":{
        "Department":"Women’s, Gender & Sexuality Study",
        "CRN":"32834",
        "Subject":"WGSS",
        "Course Number":"2120",
        "Course Name":"Introduction to Women’s, Gender, and Sexuality Stidies",
        "Tags":["Gender Culture"]
    },
    "QSS 6000":{
        "Department":"Quantitative Social Science",
        "CRN":"38119",
        "Subject":"QSS",
        "Course Number":"6000",
        "Course Name":"Quantitative Social Science",
        "Tags":["Quantitative Social Science"]
    },
    "QSS 6002":{
        "Department":"Quantitative Social Science",
        "CRN":"38120",
        "Subject":"QSS",
        "Course Number":"6002",
        "Course Name":"Probability and Statistical Modeling",
        "Tags":["Quantitative Social Science"]
    },
    "PSC 8101":{
        "Department":"Quantitative Social Science",
        "CRN":"35045",
        "Subject":"PSC",
        "Course Number":"8101",
        "Course Name":"Introduction to Empirical Political Analysis",
        "Tags":["Quantitative Social Science"]
    },
    "SEAS 0920":{
        "Department":"School of Eng & Applied Sci",
        "CRN":"32214",
        "Subject":"SEAS",
        "Course Number":"0920",
        "Course Name":"Introduction to Empirical Political Analysis",
        "Tags":["Engineering & Applied Science"]
    },
    "SEAS 0930":{
        "Department":"School of Eng & Applied Sci",
        "CRN":"32215",
        "Subject":"SEAS",
        "Course Number":"0930",
        "Course Name":"Examination Preparation",
        "Tags":["Engineering & Applied Science"]
    },
    "SEAS 0940":{
        "Department":"School of Eng & Applied Sci",
        "CRN":"31574",
        "Subject":"SEAS",
        "Course Number":"0940",
        "Course Name":"Continuing Research - Eoctoral",
        "Tags":["Engineering & Applied Science"]
    },
    "REL 1000":{
        "Department":"Religion",
        "CRN":"35769",
        "Subject":"REL",
        "Course Number":"1000",
        "Course Name":"Islam in the Digital Age",
        "Tags":["Religion Islam"]
    },
    "WGSS 1000":{
        "Department":"Religion",
        "CRN":"35769",
        "Subject":"WGSS",
        "Course Number":"1000",
        "Course Name":"Islam in the Digital Age",
        "Tags":["Religion Islam"]
    },
    "REL 1003":{
        "Department":"Religion",
        "CRN":"34949",
        "Subject":"REL",
        "Course Number":"1003",
        "Course Name":"Introduction to World Religions",
        "Tags":["Religion"]
    },
    "REL 2301":{
        "Department":"Religion",
        "CRN":"36165",
        "Subject":"REL",
        "Course Number":"2301",
        "Course Name":"Christianity",
        "Tags":["Religion Christianity"]
    },
    "REL 2401":{
        "Department":"Religion",
        "CRN":"32191",
        "Subject":"REL",
        "Course Number":"2401",
        "Course Name":"Islam",
        "Tags":["Religion Islam"]
    },
    "TSTD 3001":{
        "Department":"Tourism",
        "CRN":"30450",
        "Subject":"TSTD",
        "Course Number":"3001",
        "Course Name":"Introduction to Tourism and Hospitality Management",
        "Tags":["Tourism"]
    },
    "TSTD 3101":{
        "Department":"Tourism",
        "CRN":"30602",
        "Subject":"TSTD",
        "Course Number":"3101",
        "Course Name":"Sport and Event Business Management",
        "Tags":["Tourism"]
    },
    "TSTD 4102":{
        "Department":"Tourism",
        "CRN":"32307",
        "Subject":"TSTD",
        "Course Number":"4102",
        "Course Name":"Practicum",
        "Tags":["Tourism"]
    },
    "TSTD 4900":{
        "Department":"Tourism",
        "CRN":"38105",
        "Subject":"TSTD",
        "Course Number":"4900",
        "Course Name":"Event and Meeting Management",
        "Tags":["Tourism"]
    },
    "TSTD 6277":{
        "Department":"Tourism",
        "CRN":"30603",
        "Subject":"TSTD",
        "Course Number":"6277",
        "Course Name":"Event and Conference Management",
        "Tags":["Tourism"]
    },
    "STAT 1051":{
        "Department":"Statistics",
        "CRN":"30402",
        "Subject":"STAT",
        "Course Number":"1051",
        "Course Name":"Introduction to Business and Economic Statistics",
        "Tags":["Statistics"]
    },
    "STAT 1053":{
        "Department":"Statistics",
        "CRN":"30403",
        "Subject":"STAT",
        "Course Number":"1053",
        "Course Name":"Introduction to Statistics in Social Science",
        "Tags":["Statistics"]
    },
    "STAT 1111":{
        "Department":"Statistics",
        "CRN":"30405",
        "Subject":"STAT",
        "Course Number":"1111",
        "Course Name":"Business and Economic Statistics I",
        "Tags":["Statistics"]
    },
    "STAT 1127":{
        "Department":"Statistics",
        "CRN":"33066",
        "Subject":"STAT",
        "Course Number":"1127",
        "Course Name":"Statistics fr the Biological Sciences",
        "Tags":["Statistics"]
    },
    "STAT 1129":{
        "Department":"Statistics",
        "CRN":"34383",
        "Subject":"STAT",
        "Course Number":"1129",
        "Course Name":"Introduction to Computing",
        "Tags":["Statistics"]
    },
    "PSYD 8201":{
        "Department":"Professional Psychology",
        "CRN":"33066",
        "Subject":"PSYD",
        "Course Number":"8201",
        "Course Name":"Psychological Assessment",
        "Tags":["Psychology"]
    },
    "PSYD 8203":{
        "Department":"Professional Psychology",
        "CRN":"31519",
        "Subject":"PSYD",
        "Course Number":"8203",
        "Course Name":"Second Year Asessment",
        "Tags":["Psychology"]
    },
    "PSYD 8204":{
        "Department":"Professional Psychology",
        "CRN":"30694",
        "Subject":"PSYD",
        "Course Number":"8204",
        "Course Name":"Biological bases of Clinical Psychology",
        "Tags":["Psychology"]
    },
    "PSYD 8205":{
        "Department":"Professional Psychology",
        "CRN":"30675",
        "Subject":"PSYD",
        "Course Number":"8205",
        "Course Name":"Psychodynamic Psychopathology",
        "Tags":["Psychology"]
    },
    "PSYD 8221":{
        "Department":"Professional Psychology",
        "CRN":"31052",
        "Subject":"PSYD",
        "Course Number":"8221",
        "Course Name":"Psychodynamic Psychotherapy II",
        "Tags":["Psychology"]
    },
    "PUBH 0920":{
        "Department":"Public Health",
        "CRN":"31996",
        "Subject":"PUBH",
        "Course Number":"0920",
        "Course Name":"Continuing Research -Master’s",
        "Tags":["Public Health"]
    },
    "PUBH 0940":{
        "Department":"Public Health",
        "CRN":"31995",
        "Subject":"PUBH",
        "Course Number":"0940",
        "Course Name":"Continuing Research- Doctoral",
        "Tags":["Public Health"]
    },
    "PUBH 1010":{
        "Department":"Public Health",
        "CRN":"38004",
        "Subject":"PUBH",
        "Course Number":"1010",
        "Course Name":"First-Year Experience in Public Health",
        "Tags":["Public Health"]
    },
    "PUBH 1101":{
        "Department":"Public Health",
        "CRN":"31076",
        "Subject":"PUBH",
        "Course Number":"1101",
        "Course Name":"Introduction to Public Health and Health Services",
        "Tags":["Public Health"]
    }

}

#Place tags here pls
all_tags = [
    "Columbian College",
    "Classical Studies",
    "Photojournalism",
    "Music",
    "Strategic Mgt & Public Policy",
    "Science",
    "Sustainability",
    "Spanish",
    "Sociology",
    "Speech, Language & Hearing Science",
    "Media&Public Affairs",
    "Forensics Practice (Debate)",
    "Russian",
    "Gender Culture",
    "Quantitative Social Science",
    "Engineering & Applied Science",
    "Religion Islam",
    "Religion",
    "Religion Christianity", 
    "Tourism",
    "Statistics",
    "Psychology",
    "Public Health"    
]
class UniTagSearch(QWidget):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("UniTag Course Search")
        self.setGeometry(100, 100, 700, 600)
        self.layout = QVBoxLayout()
        self.welcome_label = QLabel("Welcome to UniTag!\nPlease select the areas you are interested in:")
        self.layout.addWidget(self.welcome_label)
        checkbox_container = QWidget()
        checkbox_layout = QVBoxLayout()
        checkbox_container.setLayout(checkbox_layout)
        self.checkboxes = []
        for tag in all_tags:
            cb = QCheckBox(tag)
            self.checkboxes.append(cb)
            checkbox_layout.addWidget(cb)
        scroll = QScrollArea()
        scroll.setWidgetResizable(True)
        scroll.setWidget(checkbox_container)
        self.layout.addWidget(scroll)
        self.search_button = QPushButton("Search")
        self.search_button.clicked.connect(self.perform_search)
        self.layout.addWidget(self.search_button)
        self.result_list = QListWidget()
        self.layout.addWidget(self.result_list)
        self.setLayout(self.layout)
    def perform_search(self):
        selected_tags = [cb.text() for cb in self.checkboxes if cb.isChecked()]

        if not selected_tags:
            matched_courses = [
                f"{code}: {info['Course Name']} ({', '.join(info['Tags'])})"
                for code, info in courses.items()
            ]
        else:
            matched_courses = []
            for code, info in courses.items():
                course_tags = info["Tags"]
                if all(tag in course_tags for tag in selected_tags):
                    matched_courses.append(
                        f"{code}: {info['Course Name']} ({', '.join(info['Tags'])})"
                    )
        self.result_list.clear()
        if matched_courses:
            self.result_list.addItems(matched_courses)
        else:
            self.result_list.addItem("No courses match the selected areas.")
if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = UniTagSearch()
    window.show()
    sys.exit(app.exec_())
```
